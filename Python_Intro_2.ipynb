```python
# Operators
# Operators are used to perform operations on variables

# Arithmetic Operators
# +  -   *    /   //    %  **

a=90
b=5

print(a+b)
print(a/b)  # In Python Division Operator Returns Value In FLOAT
print(a//b) # // FLOOR Division Operator Returns Value In INT
print(a%b)  # gives remainder , modulo devide result will give result 0(for even number), 1(for odd number)
print(a**b) # a raise to the power of b
```

```python
# Comparison Operator
# <  >   <=   >=   ==   !=
# Returns Value In Boolean [True / False]

print(90<100)
print(100>300)
print(800==800)

age = 17
print(age>=18)  #  age>18  age==18

print(90!=100)
```

```python
# Logical Operators
# AND OR NOT

# AND operator requires more than one condition
# AND operator returns TRUE only when both the conditions are TRUE

print(100==100 and 50<100)
#      cond1        cond2
#       true         true


print(100==100 and 50>100)
#      cond1        cond2
#       true         false

print(10==100 and 50>100)
#      cond1        cond2
#      false        false
```

```python
# OR operator requires more than one condition
# OR operator returns TRUE when both the conditions are TRUE
# OR operator returns TRUE if any one of the conditions is TRUE
# OR operator returns FALSE when both the conditions are FALSE

print(100==100 or 50<100)
#      cond1        cond2
#       true         true


print(100==100 or 50>100)
#      cond1        cond2
#       true        false

print(10==100 or 50>100)
#      cond1        cond2
#      false        false
```

```python
# NOT operator
# inverse the result
# True --> False     False --> True

print(not(100==100 and 50<100))
```

```python
# Membership Operator
# in  not in

name = "John Smith"
print('o' in name)
print('Smith' in name)
print('s' in name)

# not in
print('J' not in name)  # 'J' does not exist in name  false
print('s' not in name)  # 's' does not exist in name  true
```
