```python
num = int(input("Enter a number: "))

# Store original number
original_num = num

# Reverse the number using while loop
reversed_num = 0
while num > 0:
    digit = num % 10
    reversed_num = reversed_num * 10 + digit
    num = num // 10

# Check if original and reversed are the same
if original_num == reversed_num:
    print("Palindrome")
else:
    print("Not a palindrome")
```

```python
import random
attempt = 0
max_attempt = 4

name = input('Tell Me Your Name: ')
print('Welcome User : ',name, '\nGame Is Loading. Please Wait...')
print('Our Machine Is Generating One Random Number B/W 1 till 15')
num = random.randint(1,15)
print('\nNumber Is Generated. Guess The Number In 4 Attempts')

while attempt < max_attempt:
  guess_num = int(input('Enter Your Number : '))
  attempt += 1

  if guess_num < num:
    print('Your Guess Is Too low. Think of Higher Number')

  if guess_num > num:
    print('Your Guess Is Too High. Think of lower Number')

  if guess_num == num:
    print(f'Congratulation! You Won The Game in {attempt} attempts')
    break

if guess_num!=num:
    print(f'Sorry! {attempt} attempts used. You lost The Game. The Number Was {num}')
```

```python
# Guess A Number Game
# Objective : Our machine will generate one random number
# we have to guess that number within 4 attempts

import random
attempt = 0
max_attempt = 4

name = input('Tell Me Your Name : ')
print('Welcome User :',name,'\nGame Is Loading. Please Wait.....')
print('Our Machine Is Generating One Random Number B/W 1 till 15')
num = random.randint(1,15)
print('\nNumber Is Generated. Guess The Number In 4 Attempts')

while attempt < max_attempt:
  guess_num = int(input('Enter Your Number : '))
  attempt += 1

  if guess_num < num:
    print('Your Guess Is Too Low. Think Of A Higher Number')

  if guess_num > num:
    print('Your Guess Is Too High. Think Of A Lower Number')

  if guess_num == num:
    print(f'Congratulations! You Won The Game In {attempt} attempts')
    break

if guess_num!=num:
  print(f'Sorry! 0 attempts left. You Lost The Game. The Number Was {num}')
```
